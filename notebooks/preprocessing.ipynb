```python
import pandas as pd
import numpy as np

# Load final dataset
data = pd.read_csv("../data/processed/final_dataset.csv")

#Converts boolean/categorical columns to numeric
bool_cols = ['gender','smoker','known_blood_sugar','known_cholesterol','under_treatment']
for col in bool_cols:
    if col == 'gender':
        data[col] = data[col].replace({'m':0,'f':1}).astype(int)
    else:
        data[col] = data[col].astype(int)


numeric_cols = ['age','bp_systolic','bp_diastolic','est_bp_systolic','est_bp_diastolic','health_condition']
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

#Compute BP errors
data['bp_sys_error'] = abs(data['est_bp_systolic'] - data['bp_systolic'])
data['bp_dia_error'] = abs(data['est_bp_diastolic'] - data['bp_diastolic'])
data['bp_error_mean'] = (data['bp_sys_error'] + data['bp_dia_error']) / 2

#Create BP awareness groups
data['bp_awareness_group'] = pd.qcut(data['bp_error_mean'], q=3, labels=["High","Medium","Low"])

#Save processed numeric dataset
processed_path = "../data/processed/final_dataset_numeric.csv"
data.to_csv(processed_path, index=False)
```
