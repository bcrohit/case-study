```python
import pandas as pd
```

```python
# To Load raw data - with na_values to convert "null" strings to NaN
file_path = "../data/view_export_messungen_2006-11-08.xls"
data = pd.read_excel(file_path, engine='xlrd', na_values=['null', 'NULL', 'Null'])
```

```python
# To Create 'age' column from birth year (using 2006 since that's the data's year)
data['age'] = 2006 - data['geburtsjahr']
```

```python
# Removes rows with null values
data = data.dropna()
```

```python
# Keep only ages 13–100 
data = data[(data['age'] >= 13) & (data['age'] <= 100)]
```

```python
#Rename columns to English
data = data.rename(columns={
    'geschlecht': 'gender',
    'raucher': 'smoker',
    'in_behandlung': 'under_treatment',
    'messwert_bp_sys': 'bp_systolic',
    'messwert_bp_dia': 'bp_diastolic',
    'blutzucker_bekannt': 'known_blood_sugar',
    'cholesterin_bekannt': 'known_cholesterol',
    'geburtsjahr': 'birth_year',
    'befinden': 'health_condition',
    'zeit': 'datetime',
    'terminal': 'terminal',
    'postleitzahl': 'postal_code',
    'gemeinde': 'town',
    'bezirk': 'district',
    'bundesland': 'state',
    'schaetzwert_bp_sys': 'est_bp_systolic',
    'schaetzwert_by_dia': 'est_bp_diastolic'
})
```

```python
#Saving final dataset
final_path = "../data/processed/final_dataset.csv"
data.to_csv(final_path, index=False)
```
